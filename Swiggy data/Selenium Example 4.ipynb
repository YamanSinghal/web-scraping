```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select
from tqdm import tqdm
driver = webdriver.Chrome("F:/chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://www.swiggy.com/restaurants")
driver.implicitly_wait(10)
```

```python
driver.find_element_by_id("location").send_keys("jaipur") #searching restaurants in jaipur
```

```python
driver.find_element_by_xpath('//*[@id="root"]/div[1]/div[1]/div/div[1]/div[1]/div/div[2]/div/div[3]/div[1]/span[2]').click()
```

```python
e = driver.find_elements_by_class_name("_1j_Yo")
```

```python
#Just to get assured that all links are obtained. 

#for i in range(10):
 #   print(e[i].get_property("href"))
```

```python
links = []
for i in range(20):
    l = e[i].get_property("href")
    links.append(l)
```

```python
len(links)
#number of restaurants for which data is scraped
```

```python
all_details  = []
```

```python
for link in tqdm(links):
    driver.get(link)
    P = driver.find_elements_by_class_name('bQEAj')
    D = driver.find_elements_by_class_name('jTy8b')
    R = driver.find_elements_by_class_name('_2SyqU')
    T = driver.find_element_by_xpath('//*[@id="root"]/div[1]/div[1]/div[1]/div[3]/div[1]/div/div[2]/div/div[1]/div/h1').text
    
    print("length is {}".format(len(P)))
    
    for i in range(len(R)):
        r = R[i].text
        p = P[i].text
        dict1 = {"Restaurant":T,
            "Dish Name":r,
                "Price of Dish":p}
        
        all_details.append(dict1)
    
    a = len(R)
    del P[:a]
    
    
    print("updated lenght is {}".format(len(P)))
    
    for i in range(len(D)):
        d = D[i].text
        pr = P[i].text
        dict2 = {"Restaurant":T,
            "Dish Name":d,
                "Price of Dish":pr}
        
        all_details.append(dict2)
        
```

```python
import pandas as pd
data = pd.DataFrame(all_details)
data
```

```python
data.to_excel("DATA.xlsx")  #write dataframe in excel
```

```python
#To sort according to price 
data = data.astype({"Price of Dish": int})
```
